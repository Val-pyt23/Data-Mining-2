## Tambahkan kode bagian (b) agar bisa menampilkan kembali kata pertama yang di lakukan one-hot encoding (hint: lakukan inverse dari hasil encode).

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

docs = 'I ate an apple'
values = np.array(docs.split())

le = LabelEncoder()
integer_encoded = le.fit_transform(values)

print('kalimat asli : ',docs)
print('kata kata : ',values)
print('hasil integer encoding : ',integer_encoded)

print('\n')

# proses inverse
one_hot_vector = np.array([1,0,0,0])
integer_code = np.argmax(one_hot_vector)
print(f'vektor one hot : {one_hot_vector}')
print(f'integer code dari argmax : {integer_code}')

print('\n')

# proses inverse transform
inverted_word = le.inverse_transform([integer_code])
print(f'hasil one hot encoding dari kata {one_hot_vector} adalah {inverted_word}')
```

## Download file tugas_text_representation.csv dari hebat e learning, kemudian lakukan CountVectorizer dan TF-IDF pada korpus tersebut. Jelaskan hasil yang didapatkan.

```python
# Install library jika belum ada
#!pip install wikipedia
!conda install -c conda-forge wikipedia

import wikipedia
import re
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Atur bahasa Wikipedia menjadi bahasa Indonesia
wikipedia.set_lang("id")

# Daftar topik yang akan di-crawl
topics = [
    "Debit", "Kredit", "Obligasi", "Keuntungan", "Laporan laba rugi",
    "Tabungan emas", "Investasi", "Inflasi", "Deposito", "IDX"
]

# Fungsi untuk mengambil konten
def crawl_wikipedia_content(topics):
    content_list = []
    for topic in topics:
        try:
            page = wikipedia.page(topic, auto_suggest=False)
            content_list.append(page.content)
            print(f"Sukses mengambil konten: {topic}")
        except Exception as e:
            print(f"Gagal mengambil konten untuk {topic}: {e}")
    return content_list

# Jalankan fungsi crawling
corpus = crawl_wikipedia_content(topics)

# Kita sekarang memiliki variabel 'corpus' yang berisi list dari 10 dokumen (artikel)
print(f"\nTotal dokumen yang berhasil diambil: {len(corpus)}")
```

```python
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer

# Corpus dari soal
text1 = ['everybody love nlp', 'nlp is cool',
         'nlp is all about helping machines process language',
         'this tutorial is on basic nlp technique']

# --- Bagian Kode Anda (untuk menghitung IDF) ---
tf = TfidfVectorizer()
txt_fitted = tf.fit(text1)

# Mengambil nama kata (fitur) dan skor IDF-nya
feature_names = txt_fitted.get_feature_names_out()
idf_scores = tf.idf_

# --- Bagian Tambahan (untuk membuat grafik) ---
print("Membuat grafik dari nilai IDF...")

# 1. Mengatur ukuran gambar agar lebih besar dan mudah dibaca
plt.figure(figsize=(12, 6))

# 2. Membuat grafik batang (bar chart)
# Sumbu X adalah kata-kata (feature_names), Sumbu Y adalah skor IDF-nya
plt.bar(feature_names, idf_scores)

# 3. Memberi judul dan label pada sumbu
plt.title("Grafik Nilai Inverse Document Frequency (IDF) per Kata", fontsize=14)
plt.xlabel("Kata-kata (Features)")
plt.ylabel("Skor IDF")

# 4. Memutar label di sumbu X agar tidak tumpang tindih
plt.xticks(rotation=45)

# 5. Merapikan layout dan menampilkan grafik
plt.tight_layout()
plt.show()
```
